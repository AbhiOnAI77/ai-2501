```python
import re #regular expression package used below to flag 

# Step A: Remove emojis
str_with_emojis = "‚≠ê‚≠ê‚≠ê‚≠ê‚≠ê Absolutely love this! üòç Exceeded all my expectations. Fast shipping, great quality, and the customer service team was super responsive when I had questions. Will definitely be ordering again! üéâ"

emoji_pattern = re.compile("["
    u"\U0001F600-\U0001F64F"  # emoticons
    u"\U0001F300-\U0001F5FF"  # symbols & pictographs
    u"\U0001F680-\U0001F6FF"  # transport & map symbols
    u"\U0001F1E0-\U0001F1FF"  # flags
    u"\U00002702-\U000027B0"
    u"\U000024C2-\U0001F251"
    "]+", flags=re.UNICODE)

cleaned_review = emoji_pattern.sub('', str_with_emojis)
cleaned_review
```

```python
# Remove Special Characters

review = "Movie was amazing! #excited @user"

cleaned_review = re.sub(r'[^a-zA-Z0-9\s]', '',review) #a-z lower case alphabets, A-Z upper case alphabets, 0-9 numbers zero to nine, \s spaces

print(cleaned_review)
```

```python
# Remove extra spaces

str = 'I liked the      movie       '
cleaned_str = re.sub(r'\s+', ' ',str).strip() #+ includes one or more occurences in space, strip() removes white space before and after the sentence
print(str)
print(cleaned_str)
```

```python
# Combine all removals in one unit: Remove HTML Tags | Remove URLs | Remove Emojis | Remove Special Characters | Remove Extra Spaces

from bs4 import BeautifulSoup

def clean_text(test_string):
    text = BeautifulSoup(test_string, 'html.parser').get_text()
    text = re.sub(r'http/S+|wwwS+','',text)
    emoji_pattern = re.compile("["
    u"\U0001F600-\U0001F64F"  # emoticons
    u"\U0001F300-\U0001F5FF"  # symbols & pictographs
    u"\U0001F680-\U0001F6FF"  # transport & map symbols
    u"\U0001F1E0-\U0001F1FF"  # flags
    u"\U00002702-\U000027B0"
    u"\U000024C2-\U0001F251"
    "]+", flags=re.UNICODE)

    text = emoji_pattern.sub(r'', text)
    text = re.sub(r'a-zA-Z0-9\s','',text)
    text = re.sub(r'\s+',' ', str).strip()

    return text

raw_text = """ <html><body><h1>Customer Review</h1></body></html> ‚≠ê‚≠ê‚≠ê‚≠ê‚≠ê Absolutely    love this! üòç Amazing product!!! Check out this link: https://www.example.com/product?id=123 Email me at: test@email.com üéâüî• Special chars: @#$%^&*()_+-=[]{}|;':",.<>?/ Extra      spaces    everywhere   ! </body></html>"""

cleaned_text = clean_text(raw_text) #this step is 'calling the function'
print(cleaned_text)

#continue from 36:41
```

```python
# Example taken - Employee Feedback Dataset

import pandas as pd 

data = {
    'employee_id': [101,102,103,104,105,106,107,108,109,110],
    'department': ['HR','Engineering',None,'Sales','Finance','IT','Operations',None,'Customer Service','Legal'],
    'feedback': ['Great work!', 'Needs improvement',None,'','Bad','Average','Good','Needs improvement',None,'Bad'],
    'rating': [5,3,None,4,2,4,5,3,None,2],
    'manager_comments':['Great job!','Needs improvement',None,'Good work!','Bad work!','N/A','Good work!','Needs improvement',None,'Not sure'],
}
df = pd.DataFrame(data)

print('Before Cleaning: ')
print(df)
print()
print('Missing values before cleaning: ', df.isna().sum())
print()
```

```python
# Feedback Column fix (except for 'Rating' column)
df['feedback'] = df['feedback'].replace(['N/A',None,''],'No feedback provided')

df['department'] = df['department'].replace(['N/A',None,''],'Unknown')

df['manager_comments'] = df['manager_comments'].replace(['N/A','Not Sure','',None],'No manager comments')

df['rating'].fillna(df['rating'].mean(),inplace=True)

print('After cleaning:')

print(df)

print(df.isna().sum())
```
