```python
from transformers import pipeline

text_generator = pipeline(
    "text-generation",
    model='distilbert/distilgpt2',
    tokenizer='distilbert/distilgpt2',
    truncation=True #this sets the word limit of generating
)

prompt = 'Today is a rainy day in Hyderabad'

generate_texts = text_generator(
    prompt,
    max_new_tokens = 100,
    num_return_sequences=3,
    no_repeat_ngram_size=2,
    top_k=50, # internally models have data in different dimensions
    temperature = 0.8 #temperature range is 0-1. If you give more than 1 -> more randomness and less creativity
)

print(generated_texts)
```

#### Question Answer

```python
from transformers import pipeline

qa_model = pipeline('question-answering', model="distilbert-base-cased-distilled-squad")

# QnA Pair One
#question = "What is the capital of France?"
#context = "The capital of France is Paris."

# QnA Pair Two
# context = ''' The first iPhone was released by Apple in 2007 and changed the smartphone industry forever.'''
# question = "When was the first iPhone released?"

# QnA Pair Three
# context = '''I am developing AI models with Python. I work as a machine learning engineer at a tech company'''
# question = "What is my job?"

# QnA Pair Four
context = '''The Eiffel Tower is located in Paris, France. It is one of the most visited monuments in the world.'''
question = "Where is the Eiffel Tower located?"

answer = qa_model(question=question, context=context)

print(answer)
```
